```python
x = df.drop("Employee_Turnover",axis = 1)
y = df["Employee_Turnover"]

x_train,x_test,y_train,y_test = train_test_split(
    x,y,test_size =0.2,random_state = 42)

y_train.head()
```

```python
mode_1 = LogisticRegression()
model_1.fit(x_train,y_train)
```

```python
model_2 = LogisticRegression(penalty='l1', solver='liblinear', C=0.5)   # L1 Lasso
model_2.fit(x_train,y_train)

model_3 = LogisticRegression(penalty='l2', C=1, max_iter=200)   # L2  Ridge
model_3.fit(x_train,y_train)
```

```python
models = {"Logistic Regression" : model_1,"L1 Lasso" : model_2,"L2 Ridge" : model_3}

for name , model in models.items():
    print(f"Model is :  {name}")
    print()

    model_pred = model.predict(x_test)

    print(f"Accuracy of model : {accuracy_score(y_test,model_pred)}")
    print()
   
    print(f"Classification report for model : {name}")
    print()
    print(classification_report(y_test,model_pred))
    print("**************************************************************************")
                
          
    
```
